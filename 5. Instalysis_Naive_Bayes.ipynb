# Multinomial Naive Bayes¬∂

The first supervised learning method we introduce is the multinomial Naive Bayes or multinomial NB model, a probabilistic learning method. The proba- bility of a document d being in class c is computed as\
\
P(c|d) ‚àù P(c) ‚àè P(tk|c) 1‚â§k‚â§nd\
                    \
where P(tk|c) is the conditional probability of term tk occurring in a docu- ment of class c.1 We interpret P(tk|c) as a measure of how much evidence tk contributes that c is the correct class. P(c) is the prior probability of a document occurring in class c. If a document‚Äôs terms do not provide clear evidence for one class versus another, we choose the one that has a higher prior probability. ‚ü®t1, t2, . . . , tnd ‚ü© are the tokens in d that are part of the vocab- ulary we use for classification and nd is the number of such tokens in d. For example, ‚ü®t1, t2, . . . , tnd ‚ü© for the one-sentence document Beijing and Taipei join the WTO might be ‚ü®Beijing, Taipei, join, WTO‚ü©, with nd = 4, if we treat the terms and and the as stop words.

```python
# building a multinomial NB model
from sklearn.naive_bayes import MultinomialNB

# instantiate NB class
mnb=MultinomialNB()

# fitting the model on training data
mnb.fit(X,Y)

# predicting probabilities of test data
mnb.predict_proba(X_testNB)
```

```python
proba=mnb.predict_proba(X_testNB)
pd.DataFrame(proba, columns=['Neutral','Negative', 'Positive'])
```

```python
mnb.score(X,Y)
```

# Bernoulli Naive Bayes\
\
An alternative to the multinomial model is the multivariate Bernoulli model or Bernoulli model. It generates an indicator for each term of the vo- cabulary, either 1 indicating presence of the term in the document or 0 indi- cating absence. The Bernoulli model has the same time complexity as the multinomial model.
The different generation models imply different estimation strategies and different classification rules. The Bernoulli model estimates PÀÜ(t|c) as the fraction of documents of class c that contain term t

```python
from sklearn.naive_bayes import BernoulliNB

# instantiating bernoulli NB class
bnb=BernoulliNB()

# fitting the model
bnb.fit(X,Y)

# predicting probability of test data
bnb.predict_proba(X_testNB)
proba_bnb=bnb.predict_proba(X_testNB)
```

```python
pd.DataFrame(proba_bnb, columns=['Neutral','Negative','Positive'])
```

```python
bnb.score(X,Y)
```
